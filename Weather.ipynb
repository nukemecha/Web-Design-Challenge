```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('Resources/cities.csv')
```

```python
df.head()
```

```python
f = df.plot.scatter(x='Lat',y='Max Temp')
```

```python
plt.figure(figsize=(10,10), dpi = 72)
plt.scatter(df['Lat'], df['Max Temp'])
plt.xlabel("Latitude")
plt.ylabel("Maximum Temperature (F)")
plt.suptitle("Maximum Temperature vs. Latitude (6/13/2018)")
plt.savefig("maxtemp")
```

```python
plt.figure(figsize=(10,10), dpi = 72)
plt.scatter(df['Lat'], df['Cloudiness'])
plt.xlabel("Latitude")
plt.ylabel("Cloudiness")
plt.suptitle("Cloudiness vs. Latitude (6/13/2018)")
plt.savefig("cloudiness")
```

```python
plt.figure(figsize=(10,10), dpi = 72)
plt.scatter(df['Lat'], df['Humidity'])
plt.xlabel("Latitude")
plt.ylabel("Humidity")
plt.suptitle("Humidity vs. Latitude (6/13/2018)")
plt.savefig("humidity")
```
